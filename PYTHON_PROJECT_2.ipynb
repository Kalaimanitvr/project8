#Task 1
#Write Python code to develop a word guessing game where employees attempt to guess a secret word from a provided list by making a series of guesses. 
#Clues, such as optional information about the number of letters or characters in the word, can be given to assist in the guessing process.


```python
import random

def get_secret_word(word_list):
    return random.choice(word_list)

def display_clue(secret_word):
    return f"The secret word has {len(secret_word)} letters."

def play_game():
    word_list = ["python", "programming", "sql", "data", "analysis"]
    secret_word = get_secret_word(word_list)
    guessed_word = ['_'] * len(secret_word)
    attempts = 1
    guessed_letters = set()

    print("Welcome to the Word Guessing Game!")
    print(display_clue(secret_word))

    while attempts > 0 and ''.join(guessed_word) != secret_word:
        print("\nGuessed word so far: " + ' '.join(guessed_word))
        guess = input("Enter a letter or guess the full word: ").strip().lower()

        if guess == secret_word:
            print(f"Congratulations! You guessed the secret word '{secret_word}'!")
            return

        elif len(guess) == 1 and guess.isalpha():
            if guess in guessed_letters:
                print(f"You've already guessed the letter '{guess}'. Try again.")
            elif guess in secret_word:
                guessed_letters.add(guess)
                for i, letter in enumerate(secret_word):
                    if letter == guess:
                        guessed_word[i] = guess
                print(f"Good guess! The letter '{guess}' is in the secret word.")
            else:
                guessed_letters.add(guess)
                attempts -= 1
                print(f"Incorrect guess. The letter '{guess}' is not in the secret word. Attempts left: {attempts}")
        else:
            print("Invalid input. Please enter a single letter or guess the full word.")

    if ''.join(guessed_word) == secret_word:
        print(f"Congratulations! You guessed the secret word '{secret_word}'!")
    else:
        print(f"Game over! The secret word was '{secret_word}'.")

if __name__ == "__main__":
    play_game()
    print("Thank you for playing!")
```

```python
#Task 2
class GameMenu:
    
    def menu():
        print("1. Play Riddle Puzzle.")
        print("2. Play Arithmetic Puzzle.")
        print("3. Play General Knowledge quiz game.")
        print("4. Play Rock-Paper-Scissor game with the computer.")
        print("5. Play Rock-Paper-Scissor game with a user.")
        print("6. Play Word-Guessing game with the computer.")
        print("Enter any option by typing a number between 1 and 6.")

    
    def play_riddle_puzzle():
        def get_riddle():
            riddles = {
                "I speak without a mouth and hear without ears. I have no body, but I come alive with wind. What am I?": "Echo",
                "You measure my life in hours, and I serve you by expiring. I die at your feet, what am I?": "Candle",
                "I have keys but no locks. I have space but no room. You can enter, but you can't go outside. What am I?": "Keyboard",
                "I am not alive, but I can grow. I don’t have lungs, but I need air. What am I?": "Fire",
                "The more of this there is, the less you see. What is it?": "Darkness"
            }
            question, answer = random.choice(list(riddles.items()))
            return question, answer

        score = 0
        total_questions = 0

        while True:
            question, answer = get_riddle()
            print("\nRiddle: ", question)
            user_answer = input("Your answer: ").strip().lower()

            if user_answer == answer.lower():
                print("Correct!")
                score += 1
            else:
                print(f"Incorrect. The correct answer is: {answer}")

            total_questions += 1

            play_again = input("Do you want to play again? (yes/no): ").strip().lower()
            if play_again != 'yes':
                break

        if total_questions > 0:
            score_percentage = (score / total_questions) * 100
            print(f"\nYour score: {score}/{total_questions} ({score_percentage:.2f}%)")
        else:
            print("No riddles answered.")

    
    def play_arithmetic_puzzle():
        def arithmetic_puzzle():
            num1 = random.randint(1, 10)
            num2 = random.randint(1, 10)
            answer = num1 + num2
            print(f"What is {num1} + {num2}?")
            user_answer = int(input("Your answer: "))
            if user_answer == answer:
                print("Correct!")
            else:
                print(f"Incorrect. The correct answer is {answer}.")

        play_again = 'yes'
        while play_again == 'yes':
            arithmetic_puzzle()
            play_again = input("Do you want to play again? (yes/no): ").strip().lower()

    
    def play_general_knowledge_quiz():
        science_questions = [
            {"question": "Which gas do plants absorb during photosynthesis?", "answer": "Carbon dioxide"},
            {"question": "What is the full name of the WWW?", "answer": "World Wide Web"},
            {"question": "How many bones are there in the adult human body?", "answer": "206"}
        ]

        mathematics_questions = [
            {"question": "What is the only even prime number?", "answer": "2"},
            {"question": "What is the sum of the first 5 natural numbers?", "answer": "15"},
            {"question": "What is the name of the mathematical symbol π?", "answer": "Pi"}
        ]

        geography_questions = [
            {"question": "What is the tallest mountain on Earth?", "answer": "Mount Everest"},
            {"question": "What is the capital of France?", "answer": "Paris"},
            {"question": "Which planet is known as the 'Red Planet'?", "answer": "Mars"}
        ]

        other_questions = [
            {"question": "Which country won the most recent FIFA World Cup (as of 2022)?", "answer": "Argentina"},
            {"question": "Who invented the light bulb?", "answer": "Thomas Edison"},
            {"question": "In what year did India declare itself a Democratic and Republic state with the adoption of the Constitution?", "answer": "1950"}
        ]

        def ask_category():
            categories = {
                "1": ("Science-based", science_questions),
                "2": ("Mathematics-based", mathematics_questions),
                "3": ("Geography-based", geography_questions),
                "4": ("Others", other_questions)
            }
            
            while True:
                print("\nChoose your category from the following list:")
                for key, (name, _) in categories.items():
                    print(f"{key}. {name}")

                choice = input("Enter the number of your chosen category: ").strip()
                
                if choice in categories:
                    return categories[choice]
                else:
                    print("Invalid choice. Please choose a valid category number.")

        def ask_question(question):
            return input(f"{question['question']} ").strip()

        def check_answer(user_answer, correct_answer):
            return user_answer.lower() == correct_answer.lower()

        def display_score(total_questions, score):
            if total_questions > 0:
                score_percentage = (score / total_questions) * 100
                print(f"\nYour score: {score}/{total_questions} ({score_percentage:.2f}%)")
            else:
                print("No questions answered.")

        def exec_questions(questions):
            score = 0
            total_questions = 0

            for question in questions:
                user_answer = ask_question(question)
                if check_answer(user_answer, question['answer']):
                    print("Correct!")
                    score += 1
                else:
                    print(f"Incorrect. The correct answer is: {question['answer']}")
                total_questions += 1

            return total_questions, score

        def play_game():
            while True:
                category_name, questions = ask_category()
                print(f"\nYou selected the category: {category_name}")
                total_questions, score = exec_questions(random.sample(questions, len(questions)))

                display_score(total_questions, score)

                play_again = input("\nDo you want to play again? (yes/no): ").strip().lower()
                if play_again != 'yes':
                    break

        play_game()

    
    def play_rock_paper_scissor_computer():
        def rock_paper_scissor():
            moves = ['rock', 'paper', 'scissors']
            computer_move = random.choice(moves)
            user_move = input("Enter your move (rock, paper, scissors): ").strip().lower()
            if user_move not in moves:
                print("Invalid move. Please enter rock, paper, or scissors.")
                return
            print(f"Computer move: {computer_move}")
            if user_move == computer_move:
                print("It's a tie!")
            elif (user_move == 'rock' and computer_move == 'scissors') or \
                 (user_move == 'paper' and computer_move == 'rock') or \
                 (user_move == 'scissors' and computer_move == 'paper'):
                print("You win!")
            else:
                print("You lose!")

        play_again = 'yes'
        while play_again == 'yes':
            rock_paper_scissor()
            play_again = input("Do you want to play again? (yes/no): ").strip().lower()

    
    def play_rock_paper_scissor_user():
        def rock_paper_scissor():
            moves = ['rock', 'paper', 'scissors']
            user1_move = input("User 1, enter your move (rock, paper, scissors): ").strip().lower()
            user2_move = input("User 2, enter your move (rock, paper, scissors): ").strip().lower()
            if user1_move not in moves or user2_move not in moves:
                print("Invalid move. Please enter rock, paper, or scissors.")
                return
            if user1_move == user2_move:
                print("It's a tie!")
            elif (user1_move == 'rock' and user2_move == 'scissors') or \
                 (user1_move == 'paper' and user2_move == 'rock') or \
                 (user1_move == 'scissors' and user2_move == 'paper'):
                print("User 1 wins!")
            else:
                print("User 2 wins!")

        play_again = 'yes'
        while play_again == 'yes':
            rock_paper_scissor()
            play_again = input("Do you want to play again? (yes/no): ").strip().lower()

    
    def play_word_guessing_game():
        def display_clue(secret_word):
            return ' '.join('_' for _ in secret_word)

        def get_secret_word(word_list):
            return random.choice(word_list)

        def play_game():
            word_list = ["python", "programming", "sql", "data", "analysis"]
            secret_word = get_secret_word(word_list)
            guessed_word = ['_'] * len(secret_word)
            attempts = 6
            guessed_letters = set()

            print("Welcome to the Word Guessing Game!")
            print(display_clue(secret_word))

            while attempts > 0 and ''.join(guessed_word) != secret_word:
                print("\nGuessed word so far: " + ' '.join(guessed_word))
                guess = input("Enter a letter or guess the full word: ").strip().lower()

                if guess == secret_word:
                    print(f"Congratulations! You guessed the secret word '{secret_word}'!")
                    return

                elif len(guess) == 1 and guess.isalpha():
                    if guess in guessed_letters:
                        print(f"You've already guessed the letter '{guess}'. Try again.")
                    elif guess in secret_word:
                        guessed_letters.add(guess)
                        for i, letter in enumerate(secret_word):
                            if letter == guess:
                                guessed_word[i] = guess
                        print(f"Good guess! The letter '{guess}' is in the secret word.")
                    else:
                        guessed_letters.add(guess)
                        attempts -= 1
                        print(f"Incorrect guess. The letter '{guess}' is not in the secret word. Attempts left: {attempts}")
                else:
                    print("Invalid input. Please enter a single letter or guess the full word.")

            if ''.join(guessed_word) == secret_word:
                print(f"Congratulations! You guessed the secret word '{secret_word}'!")
            else:
                print(f"Game over! The secret word was '{secret_word}'.")

        play_game()

def main():
    while True:
        GameMenu.menu()
        try:
            choice = int(input("Enter your choice: "))
            if choice == 1:
                GameMenu.play_riddle_puzzle()
            elif choice == 2:
                GameMenu.play_arithmetic_puzzle()
            elif choice == 3:
                GameMenu.play_general_knowledge_quiz()
            elif choice == 4:
                GameMenu.play_rock_paper_scissor_computer()
            elif choice == 5:
                GameMenu.play_rock_paper_scissor_user()
            elif choice == 6:
                GameMenu.play_word_guessing_game()
            else:
                print("You have entered the wrong option.")
        except ValueError:
            print("Invalid input! Please enter a number between 1 and 6.")
        
        repeat = input("Do you want to play another game? (Y/y for Yes, N/n for No): ")
        if repeat.lower() != 'y':
            print("Exiting the program. Goodbye!")
            break

if __name__ == "__main__":
    main()
```
